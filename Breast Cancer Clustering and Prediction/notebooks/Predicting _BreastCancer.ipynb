# Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.compose import ColumnTransformer
```

## Model A (Original data) 

```python
X= df.drop(columns=["diagnosis"])
Y= df["diagnosis"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, random_state= 123, test_size= 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.svm import SVC
svm = SVC(kernel="rbf", probability=True, random_state=42)
svm.fit(X_train, Y_train)
```

```python
y_pred_svm= svm.predict(X_test)
y_pred_proba_svm= svm.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
print(classification_report(Y_test,y_pred_svm))
print(f"ROC-AUC score:", roc_auc_score(Y_test, y_pred_proba_svm))
```

```python
cm= confusion_matrix(Y_test, y_pred_svm)
disp=ConfusionMatrixDisplay(cm)
disp.plot()
```

# Model 2 (PCA-reduced data)

```python
X2= df_pca.drop(columns=["diagnosis", "cluster"])
Y2= df_pca["diagnosis"]
```

```python
X2_train, X2_test, Y2_train, Y2_test = train_test_split(X2, Y2, random_state= 123, test_size=0.2)
```

```python
print(X2_train.shape)
print(X2_test.shape)
print(Y2_train.shape)
print(Y2_test.shape)
```

```python
svm2 = SVC(kernel="rbf", probability=True, random_state=42)
svm2.fit(X2_train, Y2_train)
```

```python
y_pred_svm2= svm2.predict(X2_test)
y_pred_proba_svm2= svm2.predict_proba(X2_test)[:,1]
```

```python
print(classification_report(Y2_test,y_pred_svm2))
print(f"ROC-AUC score:", roc_auc_score(Y2_test, y_pred_proba_svm2))
```

```python
cm2= confusion_matrix(Y2_test, y_pred_svm2)
disp2=ConfusionMatrixDisplay(cm2)
disp2.plot()
```
